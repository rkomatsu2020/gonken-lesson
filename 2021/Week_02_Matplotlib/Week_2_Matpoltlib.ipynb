# 1_1 A simple sin-curve

## Key functions:

1. plt.plot(x,y)

2. plt.show()

## Explanation:

1. In this example, we want to draw the function y=sin(x).

2. Assume we have 256 data samples for x and y. 

3. We use plt.plot(x,y) to draw the curve.

4. We plt.show() to output the curve on the Screen.

```python
import math
import matplotlib.pyplot as plt
```

```python
import math
import matplotlib.pyplot as plt

nbsample=256
xrange=(-math.pi,math.pi)
x=[]
y=[]

for n in range(nbsample):
    ratio=(n)/nbsample
    x.append(xrange[0]+(xrange[1]-xrange[0])*ratio) #************************************
    y.append(math.sin(x[-1])) #-1 means last number in x #************************************

plt.plot(x,y)
plt.show()
```

# 1_3 A simple sin-curve_line arrtibute

## Explanation:

1. In key statement 1, we show our data by using "line".

2. In key statement 1, we show our data by using "marker". 

3. The arrtibute parameter for "line" (color, linewidth, linestyle) and "marker"(markersize, markerfacecolor markeredgecolor) are different.

# 1_4 A simple sin-curve_coordinate
## Key functions:

1. matplotlib.pyplot.text(x, y, s, fontdict=None, withdash=<deprecated parameter>, **kwargs)
    
## Explanation:

1. In this example, we will learn how to add the coordinate information to our curve.

3. plt.text() ia a funciton to add text to the axes.

   matplotlib.pyplot.text(x, y, s, fontdict=None, withdash=<deprecated parameter>, **kwargs)

# 1_9 A simple sin-multi subplot
## Key statements:

1. fig=plt.figure()

2. sub_fig1=fig.add_subplot(211)
   
   sub_fig2=fig.add_subplot(212)

## Explanation:

1. To crete multi subplot, the first step is to build a Figure by using "fig=plt.figure()"

2. Then we add a new subplot on this figure, by using statement:"sub_fig1=fig.add_subplot(211)". Where 211 means, we build a in total 2*1 size of subplots(like a 2*1 matrix), and this plor is the 1st plot. 

3. Then, plot the curve.

4. Then, next subplot by using sub_fig2=fig.add_subplot(212). Where 212 means, we build a in total 2*1 size of subplots(like a 2*1 matrix), and this plor is the 2nd plot. 




```python
import math
import matplotlib.pyplot as plt
import numpy as np


nbsample=256
x=np.linspace(-math.pi,math.pi,num=nbsample)

y1=np.sin(x)
y2=np.cos(x)

fig=plt.figure()#************************************
sub_fig1=fig.add_subplot(211)#************************************
sub_fig1.grid(True)

plt.plot(x,y1,'bo')

sub_fig2=fig.add_subplot(212)#************************************
sub_fig2.grid(True)
plt.plot(x,y2,"k")
plt.show()

plt.savefig('mult_lines.png',dpi=600)
```

# 1_10 A simple sin-multi subplot_another way
## Key statements:

1. plt.subplot(2,1,1)

## Explanation:

1. We can also draw multi subplot by using "plt.subplot(2,1,1)", similarly, (2,1,1) means 1st subplot on the 2*1 size of subplots set. 

```python
import math
import matplotlib.pyplot as plt
import numpy as np


nbsample=256
x=np.linspace(-math.pi,math.pi,num=nbsample)

y1=np.sin(x)
y2=np.cos(x)

plt.subplot(2,1,1)#************************************
plt.plot(x,y1,"r-.")
plt.grid(b=True)
plt.subplot(2,1,2)#************************************
plt.plot(x,y2,"bo")
plt.grid(b=True)
plt.show()

plt.savefig('mult_lines.png',dpi=600)
```

# 2_2_3 Bar_MultiBar_1

## Key Statements:

1. a=plt.bar(y_pos,performance1,bar_width,color='b')

   b=plt.bar(y_pos+bar_width,performance2,bar_width,color='r')
   
2. plt.xticks(y_pos+bar_width/2,objects)

## Explanation:

1. We difine the width of the bar by setting parameter width in the plt.bar()

2. We plot two bards, a and b where the start coordinate of b on x-axis has length different of bar_width between the start coordinate of a, so we can put two barsagjoin each other.

3. plt.xtick has two key parameters, plt.xtick(x coordinate, name). Here, we set "x coordinate"="y_pos+bar_width" so that the x-axis name can be exact in the middle of two bar. 

```python
import math
import matplotlib.pyplot as plt

objects=('Python','Java','Matlab')
y_pos=np.arange(len(objects))

performance1=[10,8,3]
performance2=[4,5,6]

bar_width=0.35#************************************

a=plt.bar(y_pos,performance1,bar_width,color='b')#************************************
b=plt.bar(y_pos+bar_width,performance2,bar_width,color='r')#************************************

plt.xticks(y_pos+bar_width/2,objects)#************************************
plt.ylabel('Uusers number')

plt.title('Language user distribution')
```

#2_6 3-D

## Key functions:

1. ax1.plot(x,y,z)

2. ax2.scatter(x,y,z,c='r')

3. ax3.plot_surface(x,y,z,cmap=plt.cm.winter)

## Explanatioon:

1. The way to draw 3-d diagram is same as drawing a 2-d diagram, the only different in 3-d, we have 3 coordinate, x, y and z. 

2. In this example, we first build a figure, andf plot the "3-d line curve" in figure position 221, "3-d scatter" in figure position 222 and "3-d surface" in fugure position 223.

```python
#!/usr/bin/env python
# coding=utf-8
import numpy as np 
import matplotlib.pyplot as plt 
#import 3D plot model
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(20,10))

#3d-Curve
ax1 = fig.add_subplot(221,projection='3d')#************************************
theta = np.linspace(-4*np.pi,4*np.pi,500)
z = np.linspace(-2,2,500)
r = z**2 + 1
x = r*np.sin(theta)
y = r*np.cos(theta)

ax1.plot(x,y,z)#************************************
ax1.set_xlabel('x',fontsize=15)
ax1.set_ylabel('y',fontsize=15)
ax1.set_zlabel('z',fontsize=15)

#3d-Scatter
ax2 = fig.add_subplot(222,projection='3d')#************************************
x = np.random.randn(500)
y = np.random.randn(500)
z = np.random.randn(500)

ax2.scatter(x,y,z,c='r')#************************************
ax2.set_xlabel('x',fontsize=15)
ax2.set_ylabel('y',fontsize=15)
ax2.set_zlabel('z',fontsize=15)

#3d-Surface
ax3 = fig.add_subplot(223,projection='3d')#************************************
x = np.linspace(-2,2,500)
y = np.linspace(-2,2,500)
x,y = np.meshgrid(x,y)
z = np.sqrt(x**2 + y**2)
ax3.plot_surface(x,y,z,cmap=plt.cm.winter)#************************************
ax3.set_xlabel('x',fontsize=15)
ax3.set_ylabel('y',fontsize=15)
ax3.set_zlabel('z',fontsize=15)

plt.show()
```

# 3_1 Pair plot

## Key functions:

1. sns.pairplot

## Explanation:

1. Pairplot: ペアプロット図（散布図行列）

2. Function: Plot pairwise relationships in a dataset.

   By default, this function will create a grid of Axes such that each numeric variable in data will by shared across the y-axes across a single row and the x-axes across a single column.
   
   The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal distribution of the data in each column.

   It is also possible to show a subset of variables or plot different variables on the rows and columns.

# 3_3 Density plot

## Key Statements:
1. sns.kdeplot(iris.loc[iris['species'] == 'Iris-versicolor', 'sepal_length'],
             shade=False, vertical = False, color="g", label="Iris-versicolor", alpha=.7)
             
## Explanation:

1. The data is still "iris.csv", which contains 3 different iris species, 'Iris-versicolor','Iris-virginica' and 'Iris-setosa'. We want to use    the density plot to find the probability density distribution of petal length for different iris species. 


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# read data
iris = pd.read_csv("iris.csv")
iris.columns=['sepal_length','sepal_width','petal_length','petal_width','species']

#plot
sns.kdeplot(iris.loc[iris['species'] == 'Iris-versicolor', 'sepal_length'],
             shade=False, vertical = False, color="g", label="Iris-versicolor", alpha=.7)

sns.kdeplot(iris.loc[iris['species'] == 'Iris-virginica', 'sepal_length'],
             shade=False, vertical = False, color="deeppink", label="Iris-virginica", alpha=.7)

sns.kdeplot(iris.loc[iris['species'] == 'Iris-setosa', 'sepal_length'],
             shade=False, vertical = False, color="dodgerblue", label="Iris-setosa", alpha=.7)

# Decoration
plt.title('Density map of iris petal length', fontsize=16)
plt.legend()
plt.show()
```

2. A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one      (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components           correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.
 
   This can be an effective and attractive way to show multiple distributions of data at once, but keep in mind that the estimation procedure       is influenced by the sample size, and violins for relatively small samples might look misleadingly smooth.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# read data from csv
iris = pd.read_csv("iris.csv")
iris.columns=['sepal_length','sepal_width','petal_length','petal_width','species']

# plot
fig, axes = plt.subplots(2, 2, figsize=(7, 5), sharex=True)

sns.violinplot(x = 'species', y = 'sepal_length', 
               data = iris, split = True, 
               scale='width', inner="box", 
               ax = axes[0, 0])
sns.violinplot(x = 'species', y = 'sepal_width', 
               data = iris, split = True, scale='width', 
               inner="box", 
               ax = axes[0, 1])
sns.violinplot(x = 'species', y = 'petal_length', 
               data = iris, split = True, scale='width', 
               inner="box", 
               ax = axes[1, 0])
sns.violinplot(x = 'species', 
               y = 'petal_width', 
               data = iris, split = True, 
               scale='width', inner="box", 
               ax = axes[1, 1])

plt.setp(axes, yticks=[])
plt.tight_layout()
```
